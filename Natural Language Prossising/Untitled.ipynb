# stopwords

```python
from nltk.corpus import stopwords
sw=set(stopwords.words('english'))
```

```python
sw
```

```python
def remove_stopwords(text,stopwords):
    useful_words=[w for w in text if w not in stopwords]
    return useful_words
```

```python
text="i am not a student i am worker".split()
type(text)
useful_text=remove_stopwords(text,sw)
print(useful_text)
```

# more way to create Features
* Unigram-every word as feature
* Bigram
* Trigram
* n-gram
* TF-IDE Normalisation

```python
sent_1=["this is good movie"]
sent_3=["this is not good movie"]
sent_2=['this is good movie but actor is not present']
```

```python
#unigram
cv=CountVectorizer()#default
docs=[sent_1[0],sent_2[0],sent_3[0]]
cv.fit_transform(docs).toarray()
# if we use unigram then 3rd review it think  as a good review
#so for that we need to use bigram(not good) as negative review
```

```python
# Bi-gram
cv=CountVectorizer(ngram_range=(2,2))
docs=[sent_1[0],sent_2[0],sent_3[0]]
cv.fit_transform(docs).toarray()
```

```python
cv.vocabulary_
```

```python
#tri-gram
cv=CountVectorizer(ngram_range=(3,3))
docs=[sent_1[0],sent_2[0],sent_3[0]]
cv.fit_transform(docs).toarray()
```

```python
cv.vocabulary_
```

```python
#n-gram
cv=CountVectorizer(ngram_range=(1,3))
docs=[sent_1[0],sent_2[0],sent_3[0]]
cv.fit_transform(docs).toarray()
```

```python
cv.vocabulary_
```

# Tf-idf Normalization
* Avoid feature that occer,very offen,becouse they contain less information


```python
sent_1='this is good movie'
sent_2='this was good movie'
sent_3='this is not good movie'

corpus=[sent_1,sent_2,sent_3]
```

```python
# tf-->tearm frequency td(tearm,document)
#idf--> inversr document frequency idf(t,d)=log(n/1+count(t,D))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer()
```

```python
vc=tfidf.fit_transform(corpus).toarray()
```

```python
print(vc)
```

```python
tfidf.vocabulary_
```
